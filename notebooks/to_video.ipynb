```python
writer = cv2.VideoWriter('trashcam.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 100, (crop[2]-crop[0],crop[3]-crop[1]), True)
t = time()
for i in range(len(imgs)):
    im = cv2.imread(str(imgs[i]))
    writer.write(im[crop[1]:crop[3],crop[0]:crop[2],:])
    if i % 100 == 0:
        print(i, time() - t, i/len(imgs)*100)
        t = time()
cv2.destroyAllWindows()
writer.release()
```
